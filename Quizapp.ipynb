```python


win=Tk()
win.state("zoomed")
win.config(bg="powder blue")


lbl_title=Label(win,text="Welcome to the quiz",font=("lucida calligraphy",36,"bold","underline"),bg="powder blue")
lbl_title.pack(pady=15)


img_signin=Image.open("signin.png").resize((140,40))
img_signintk=ImageTk.PhotoImage(img_signin)

img_signup=Image.open("signup.png").resize((140,40))
img_signuptk=ImageTk.PhotoImage(img_signup)

img_sub=Image.open("submit.png").resize((140,40))
img_subtk=ImageTk.PhotoImage(img_sub)

img_back=Image.open("back.jfif").resize((100,30))
img_backtk=ImageTk.PhotoImage(img_back)

questions=[
    "MS-Word is an example of-",
    "A computer cannot boot if it does not have the _____",
    "________ is the process of dividing the disk into tracks and sectors",
    "Junk e-mail is also called ______",
    "Microsoft Office is an example of a",
    "By default, your documents print in ________ mode",
    "_____are attempts by individuals to obtain confidential information from you by falsifying their identity",
    "Storage capacity of magnetic disk depends on",
    "Where is RAM located?",
    "A _______ is a software program used to view Web pages.",
        
    ]

answers_choice=[
    ["An operating system","A processing device","Application software","An input device",],
    ["Compiler","Loader","Operating system","Assembler",],
    [" Tracking"," Formatting","Crashing","Allotting",],
    ["Spam"," Spoof","Sniffer script"," Spool",],
    [" Closed source software","Open source software","Horizontal market software","vertical market software",],
    [" Landscape","Portrait","Page Setup","Print View",],
    ["Phishing trips","Computer viruses","Phishing scams"," Spyware scams",],
    ["disk pack in disk surface","tracks per inch of surface","bits per inch of tracks"," All of the above",],
    ["Mother Board","Expansion Board","External Drive","None",],
    ["site"," host"," link","browser",],
    ]

answers=[2,2,1,0,2,1,2,3,0,2]
  
indexes=[]

user_answer=[]



def gen():
    global indexes
    while(len(indexes)<5):
        x=random.randint(0,9)
        if x in indexes:
            continue
        else:
            indexes.append(x)
            
def showresult(score):
    global frm1
    frm1.destroy()
    lbl_questions.destroy()
    r1.destroy()
    r2.destroy()
    r3.destroy()
    r4.destroy()
    lbl_title.config(text="Result")

    labelimage=Label(win,background="powder blue")
    labelimage.pack(pady=60)
    label_resulttext=Label(win,font=("lucida calligraphy",20),bg="powder blue")
    label_resulttext.pack()
    if score>=20:
        img=PhotoImage(file="welldone.png")
        labelimage.configure(image=img)
        labelimage.image=img
        label_resulttext.configure(text="Grade-A")
    elif (score>=10 and score<=20):
        img=PhotoImage(file="nicetry.png").subsample(4,4)
        labelimage.configure(image=img)
        labelimage.image=img
        label_resulttext.configure(text="Grade-B")
    else:
        img=PhotoImage(file="revise1.png").subsample(2,2)
        labelimage.configure(image=img,bg="powder blue")
        labelimage.image=img
        label_resulttext.configure(text="Grade-C")
        
        
               
            
def calc():
    global indexes,user_answer,answer
    score=0
    x=0
    for i in indexes:
        if user_answer[x]==answers[i]:
            score=score+5
        x=x+1
    showresult(score)    

ques=1    
def selected():
    global radiovar,user_answer
    global lbl_questions,r1,r2,r3,r4
    global ques
    x=radiovar.get()
    user_answer.append(x)
    radiovar.set(-1)
    if ques<5:
        lbl_questions.config(text=questions[indexes[ques]])
        r1['text']=answers_choice[indexes[ques]][0]
        r2['text']=answers_choice[indexes[ques]][1]
        r3['text']=answers_choice[indexes[ques]][2]
        r4['text']=answers_choice[indexes[ques]][3]
        ques=ques+1
        
    else:
        calc()
        
    

def home_frame():  

    frm=Frame(win,highlightthickness=4,highlightbackground="blue",background="LavenderBlush2")
    frm.place(x=330,y=180,relwidth=.45,relheight=.5)
 

    def signin(event):
        global loginuser_row
        u=e_username.get()
        p=e_password.get()
        
        
        
        
        con=sqlite3.connect(database="quizquestions.sqlite")
        cursor=con.cursor()
        cursor.execute("select * from users where username=? and password=?",(u,p))
        loginuser_row=cursor.fetchone()
        con.close()
        if(loginuser_row==None):
            
            messagebox.showerror("Login","Invalid Username/Password")
        else:
            frm.destroy()
            startquiz_frame()
            
            
    def signup(event):
        frm.destroy()
        newuser_frm()
         
    
    lbl_username=Label(frm,text="Username",font=("",18,"bold"),background="LavenderBlush2")
    lbl_username.place(x=120,y=50)
    e_username=Entry(frm,font=("",18,"bold"),width=12)
    e_username.focus()
    e_username.place(x=260,y=50)

    lbl_password=Label(frm,text="Password",font=("",18,"bold"),background="LavenderBlush2")
    lbl_password.place(x=120,y=130)
    e_password=Entry(frm,font=("",18,"bold"),width=12,show="*")
    e_password.place(x=260,y=130)

    lbl_btn_signin=Label(frm,image=img_signintk,background="LavenderBlush2")
    lbl_btn_signin.place(x=120,y=220)
    lbl_btn_signin.bind("<Button>",signin)
    
    
    lbl_btn_signup=Label(frm,image=img_signuptk,background="LavenderBlush2")
    lbl_btn_signup.place(x=290,y=220)
    lbl_btn_signup.bind("<Button>",signup)

def newuser_frm():
            
    frm=Frame(win,highlightthickness=4,highlightbackground="red",background="LavenderBlush2")
    frm.place(x=330,y=180,relwidth=.45,relheight=.6)
    
    lbl_title.config(text="Register yourself")
    
    lbl_fullname=Label(frm,text="Your Name",font=("",15,"bold"),background="LavenderBlush2")
    lbl_fullname.place(x=100,y=50)
    e_fullname=Entry(frm,font=("",15,"bold"),width=15)
    e_fullname.focus()
    e_fullname.place(x=240,y=50)
    
    lbl_username=Label(frm,text="Username",font=("",15,"bold"),background="LavenderBlush2")
    lbl_username.place(x=100,y=120)
    e_username=Entry(frm,font=("",15,"bold"),width=15)
    e_username.place(x=240,y=120)
    
    lbl_password=Label(frm,text="Password",font=("",15,"bold"),background="LavenderBlush2")
    lbl_password.place(x=100,y=190)
    e_password=Entry(frm,font=("",15,"bold"),width=15,show="*")
    e_password.place(x=240,y=190)
    
    lbl_location=Label(frm,text="Location",font=("",15,"bold"),background="LavenderBlush2")
    lbl_location.place(x=100,y=260)
    e_location=Entry(frm,font=("",15,"bold"),width=15)
    e_location.place(x=240,y=260)
    
    def back(event):
        frm.destroy()
        home_frame()
    
    def register_db(event):
        fullname=e_fullname.get()
        username=e_username.get()
        password=e_password.get()
        location=e_location.get()
        
        if(len(fullname)==0 or len(username)==0 or len(password)==0 or len(location)==0):
            messagebox.showwarning("Validation","Please fill all fields!")
        else:
            con=sqlite3.connect(database="quizquestions.sqlite")
            cursor=con.cursor()
            cursor.execute("insert into users values(?,?,?,?)",(fullname,username,password,location))
            con.commit()
            con.close()
            messagebox.showinfo("Registration","registration done!")
            frm.destroy()
            home_frame()
            
    
            
    
    lbl_btn_submit=Label(frm,image=img_subtk,background="LavenderBlush2")
    lbl_btn_submit.place(x=200,y=320)
    lbl_btn_submit.bind("<Button>",register_db)
    
    
    lbl_btn_back=Label(frm,image=img_backtk,background="LavenderBlush2")
    lbl_btn_back.place(x=0,y=0)
    lbl_btn_back.bind("<Button>",back)
    
    
def startquiz_frame():
    global radiovar,user_answer,frm1
    global lbl_questions,r1,r2,r3,r4
    gen()
    #playsound("gypsy.mp3")
    
    lbl_title.config(text="Lets, begin the Quiz")
    
    frm1=Frame(win,highlightthickness=4,highlightbackground="blue",background="LavenderBlush2")
    frm1.place(x=350,y=160,relwidth=.45,relheight=.6)
    
    lbl_questions=Label(frm1,text=questions[indexes[0]],font=("Constantia",16),width=500,justify="center",wraplength=400,background="LavenderBlush2")
    lbl_questions.pack(pady=60)
    
    radiovar= IntVar()
    radiovar.set(-1)
    
    r1=Radiobutton(frm1,text=answers_choice[indexes[0]][0],font=("Constantia",16),value=0,variable=radiovar,command=selected,background="LavenderBlush2")
    r1.pack(pady=4)                       
    r2=Radiobutton(frm1,text=answers_choice[indexes[0]][1],font=("Constantia",16),value=1,variable=radiovar,command=selected,background="LavenderBlush2") 
    r2.pack(pady=4)                       
    r3=Radiobutton(frm1,text=answers_choice[indexes[0]][2],font=("Constantia",16),value=2,variable=radiovar,command=selected,background="LavenderBlush2")
    r3.pack(pady=4)                      
    r4=Radiobutton(frm1,text=answers_choice[indexes[0]][3],font=("Constantia",16),value=3,variable=radiovar,command=selected,background="LavenderBlush2")
    r4.pack(pady=4)                       
            

home_frame()
win.mainloop()

```
